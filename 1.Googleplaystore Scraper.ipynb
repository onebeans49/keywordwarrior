```python
from google_play_scraper import Sort, reviews, search
import pandas as pd
import os
import csv

# Google Play Store ë¦¬ë·° ìŠ¤í¬ë˜í•‘ í•¨ìˆ˜
def scrap_reviews(app_list, num_reviews_per_app, total_reviews, save_dir):
    """
    ì£¼ì–´ì§„ ì•± ë¦¬ìŠ¤íŠ¸ë¡œë¶€í„° ë¦¬ë·°ë¥¼ ìˆ˜ì§‘í•˜ê³ , CSV íŒŒì¼ë¡œ ì €ì¥í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.

    ë§¤ê°œë³€ìˆ˜:
    app_list (list): ë¦¬ë·°ë¥¼ ìˆ˜ì§‘í•  ì•±ì˜ ì´ë¦„ë“¤ì„ ë‹´ì€ ë¦¬ìŠ¤íŠ¸ì…ë‹ˆë‹¤.
    num_reviews_per_app (int): ê° ì•±ë§ˆë‹¤ ìˆ˜ì§‘í•  ë¦¬ë·°ì˜ ìˆ˜ì…ë‹ˆë‹¤.
    total_reviews (int): ì´ ìˆ˜ì§‘í•  ë¦¬ë·°ì˜ ìˆ˜ì…ë‹ˆë‹¤.
    save_dir (str): ë¦¬ë·°ë¥¼ ì €ì¥í•  ë””ë ‰í† ë¦¬ì˜ ê²½ë¡œì…ë‹ˆë‹¤.

    ë°˜í™˜ê°’:
    None
    """
    total_count = 0  # ì´ ë¦¬ë·° ìˆ˜ë¥¼ ì¶”ì í•˜ê¸° ìœ„í•œ ì¹´ìš´í„°ì…ë‹ˆë‹¤.

    for num, app_name in enumerate(app_list):
        # í•´ë‹¹ ì•±ì˜ ê²€ìƒ‰ ê²°ê³¼ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.
        search_results = search(app_name, lang="ko", country="kr", n_hits=1)
        app_id = search_results[0]['appId']

        # ì•± IDë¥¼ ì‚¬ìš©í•˜ì—¬ í•œêµ­ì–´ ë¦¬ë·°ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.
        reviews_list = []
        continuation_token = None

        while len(reviews_list) < num_reviews_per_app:  # ì›í•˜ëŠ” ë¦¬ë·° ìˆ˜ì— ë„ë‹¬í•  ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.
            result, continuation_token = reviews(
                app_id,
                count=min(50, num_reviews_per_app - len(reviews_list)),  # í•„ìš”í•œ ë§Œí¼ë§Œ ë¦¬ë·°ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.
                lang='ko',
                country='kr',
                sort=Sort.NEWEST,
                continuation_token=continuation_token
            )

            # ë§Œì•½ ë” ì´ìƒ ë¦¬ë·°ê°€ ì—†ë‹¤ë©´ ë°˜ë³µë¬¸ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.
            if not result:
                break

            reviews_list.extend(result)

        # ê°€ì ¸ì˜¨ ë¦¬ë·°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜í•˜ê³  CSV íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.
        df = pd.DataFrame(reviews_list)
        df.to_csv(os.path.join(save_dir, f'{app_name}.csv'), index=False, encoding='utf-8-sig', escapechar='\\', quoting=csv.QUOTE_ALL)

        # ê° ì•±ì— ëŒ€í•´ ìˆ˜ì§‘ëœ ë¦¬ë·° ìˆ˜ë¥¼ ì¶œë ¥
        print(f"Collected {len(reviews_list)} reviews for '{app_name}'")

        # ì´ ë¦¬ë·° ìˆ˜ë¥¼ ì—…ë°ì´íŠ¸í•˜ê³ , ì›í•˜ëŠ” ì´ ë¦¬ë·° ìˆ˜ì— ë„ë‹¬í–ˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.
        total_count += len(df)
        if total_count >= total_reviews:
            break

    # ëª¨ë“  ë¦¬ë·°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜í•˜ê³  CSV íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.
    df_all = pd.DataFrame(reviews_list)
    df_all.insert(0, 'game', continuation_token)
    df_all.to_csv(f'{save_dir}/all_reviews.csv', index=False, encoding='utf-8-sig', escapechar='\\', quoting=csv.QUOTE_ALL)


    # ì´ ë¦¬ë·° ìˆ˜ê°€ ì›í•˜ëŠ” ìˆ˜ì— ë„ë‹¬í•˜ì§€ ì•Šì€ ê²½ìš° ê²½ê³  ë©”ì‹œì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
    if total_count < total_reviews:
        print(f'Warning: The total number of reviews gathered ({total_count}) is less than the desired total ({total_reviews}).')
    print(total_count,'ê°œ ìˆ˜ì§‘')
```
